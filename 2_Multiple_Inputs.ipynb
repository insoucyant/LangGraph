```python
from typing import TypedDict, List 
from langgraph.graph import StateGraph
```

```python
# Implement a simple state graph with multiple inputs
# Implement StateSchema first
class AgentState(TypedDict):
    values: List[int]
    name: str
    result: str
```

```python
# Build Node
def process_inputs(state: AgentState) -> AgentState:
    ### This node processes the multiple different inputs and updates the state ###
    print(f"Processing inputs: {state}")
    state['result'] = f"Processed {state['name']}'s states with values {sum(state['values'])}"
    print(f"Updated state: {state}")
    return state
```

```python
# Create the graph
# StateGraph to hold the state and nodes. It will manage the flow of data through the nodes.
# StateGraph initializes the graph and defines the initial state as per the state schema: AgentState with default values.
graph = StateGraph(AgentState)
```

```python
graph.add_node("InputProcessor", process_inputs)
graph.set_entry_point("InputProcessor")
graph.set_finish_point("InputProcessor")
app = graph.compile()
```
